```python
# Import the csv dataset
nRowsRead  = 1000

stats = pd.read_csv('../../raw_data/age_standardized_suicide_rates.csv',delimiter =',', nrows = nRowsRead)

stats.dataframeName = 'age_standardized_suicide_rates.csv'
nRow, nCol = stats.shape
print(f' There are {nRow} rows and {nCol} colums')

```

```python
# Ploting different years sucide rates for different sex groups for all the countries
sns.set_style("darkgrid")

f, axes = plt.subplots(2, 2, figsize=(15,15))

vis1 = sns.distplot(stats["2016"],ax=axes[0,0])
vis1.set_title('Comparison of Suicide rate in different years')
vis1.set_ylabel('Suicide rate average 100,000 population')

vis2 = sns.distplot(stats["2015"],ax=axes[0,1])
vis2.set_title('Comparison of Suicide rate in different years')
vis2.set_ylabel('Suicide rate average per 100,000 population')

vis3 = sns.distplot(stats["2010"],ax=axes[1,0])
vis3.set_title('Comparison of Suicide rate in different years')
vis3.set_ylabel('Suicide rate average per 100,000 population')

vis4 = sns.distplot(stats["2000"],ax=axes[1,1])
vis4.set_title('Comparison of Suicide rate in different years')
vis4.set_ylabel('Suicide rate average per 100,000 population')

```
