```python
# Import python packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']= 8,4
import warnings
warnings.filterwarnings('ignore')

```

```python
# Import the csv dataset
crude = pd.read_csv('../../raw_data/crude_suicide_rates.csv')
```

```python
# changeing Country column type from object to category

crude.Country = crude.Country.astype('category')
```

```python
# changeing Sex column type from object to category

crude.Sex = crude.Sex.astype('category')
```

```python
# Creating scatter plot on different age groups

plt.scatter(crude[' 80_above'], crude[' 70to79'])

plt.scatter(crude[' 60to69 '], crude[' 50to59 '])

plt.scatter(crude[' 40to49'], crude[' 30to39'])

plt.scatter(crude[' 20to29'], crude[' 10to19'])

```

```python
# Ploting different aged group suicide rates in boxplot

sns.set_style("darkgrid")
f, axes = plt.subplots(2, 4, figsize=(15,15))

z = sns.boxplot(data=crude, x='Sex', y=' 80_above', ax=axes[0,0])
z1 = sns.boxplot(data=crude, x='Sex', y= ' 70to79', ax=axes[0,1])
z2 = sns.boxplot(data=crude, x='Sex', y=' 60to69 ', ax=axes[0,2])
z3 = sns.boxplot(data=crude, x='Sex', y=' 50to59 ', ax=axes[0,3])
z4 = sns.boxplot(data=crude, x='Sex', y=' 40to49', ax=axes[1,0])
z5 = sns.boxplot(data=crude, x='Sex', y=' 30to39', ax=axes[1,1])
z6 = sns.boxplot(data=crude, x='Sex', y=' 20to29', ax=axes[1,2])
z7 = sns.boxplot(data=crude, x='Sex', y=' 10to19', ax=axes[1,3])
```
