```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gini = pd.read_csv('../raw_data/economic-inequality-gini-index.csv',
                  names = ['country','country_code','year','gini_coef'],
                  skiprows = 1)
gini
```

```python
suiciderate = pd.read_csv('../raw_data/crude_suicide_rates.csv',
                         names = ['country','sex','>80','70-79','60-69','50-59','40-49','30-39','20-29','10-19'],
                         skiprows = 1)
suiciderate = suiciderate.assign(working_age = lambda x: x['50-59']+x['40-49']+x['30-39']+x['20-29'],
                                 sex = lambda x: x['sex'].str.strip(' '))
suiciderate
```

```python
suiciderate_bothsexes = suiciderate[suiciderate['sex'] == 'Both sexes']
suiciderate_bothsexes
```

```python
suiciderate_gini = suiciderate_bothsexes.merge(gini, how = 'inner', on='country')
suiciderate_gini
```

```python
sns.set_theme(style='white')

sns.regplot(data = suiciderate_gini, x = 'working_age', y = 'gini_coef', scatter_kws={"s": 80})
plt.xlim(0, None)
plt.ylim(20, None)

plt.show
```
